```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.arange(0,10)
y = x ** 3
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'*')
```

```python
plt.plot(x,y,'r-')
```

```python
plt.plot(x,y,'r--')
```

```python
plt.plot(x,y,'r--')
plt.xlim(0,5) # limits till 5
plt.ylim(0,100) # limits till 100
plt.title('Graph with x and y limited to a certain range')
plt.xlabel('Indepdent variable X')
plt.ylabel('Dependent variable')
```
